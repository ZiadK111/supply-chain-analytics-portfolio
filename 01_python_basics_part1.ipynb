## Buliding a Gessing Game

* The game is that we will have a secret word. Then the user will try to guess this secret word.

```python
secret_word = "giraffe"
guess = ""
guess_count = 0
guess_limit = 3
out_of_guesses = False

while guess != secret_word and not(out_of_guesses):
  if guess_count < guess_limit:
    guess = input("Enter guess:")
    guess_count += 1
  else:
    out_of_guesses = True

if out_of_guesses:
  print("Out of guesses, you lose!")
else:
  print("You Win!")
```

# Procurement Practice - Using For Loops for Procurement Data


```python
items = [
    {"name": "Laptop", "quantity": 8, "price": 1250},
    {"name": "Monitor", "quantity": 15, "price": 320},
    {"name": "Keyboard", "quantity": 30, "price": 45},
    {"name": "Mouse", "quantity": 50, "price": 25}
]

total_quantity = 0
for item in items:
    total_quantity += item["quantity"]

print("Total Quantity:", total_quantity)

total_price = 0
for item in items:
    total_price += item["price"]

average_price = total_price / len(items)
print("Average Price per Item:", round(average_price, 2))
```

### Thought Process Behind This Code

The goal of this code is to analyze a small set of procurement data to derive some basic insights: the total number of items and the average cost per item. Here's how one might think about approaching this task:

1.  **Understanding the Data:** We have several different types of items, each with a name, a quantity purchased, and a unit price. This suggests a structured way to store each item's information.

2.  **Choosing a Data Structure:**
    *   Since each item has multiple attributes (name, quantity, price), a **dictionary** is a good choice to represent a single item (e.g., `{"name": "Laptop", "quantity": 8, "price": 1250}`).
    *   Since we have *multiple* items, a **list** is suitable to hold all these individual item dictionaries (e.g., `[item1, item2, item3]`).

3.  **Calculating Total Quantity:**
    *   To find the total quantity, we need to go through each item in our list.
    *   For each item, we need to access its `"quantity"` value.
    *   We need a running total, so we'll start with `total_quantity = 0` and add each item's quantity to it.
    *   A **`for` loop** is the perfect construct to iterate over each item in the list.

4.  **Calculating Average Price Per Item:**
    *   To find the *average price per item*, we first need the *total sum of individual prices* of each unique type of item (e.g., Laptop's price + Monitor's price + Keyboard's price + Mouse's price).
    *   Similar to quantity, we'll initialize `total_price = 0`.
    *   Again, a **`for` loop** is needed to iterate through each item.
    *   Inside the loop, we access the `"price"` value of each item and add it to `total_price`.
    *   Once we have the `total_price`, we need to divide it by the *number of different item types* (not the total quantity of all items combined, as that would give a different metric). The number of different item types is simply the length of our `items` list, which can be found using `len(items)`.
    *   It's good practice to round monetary values, so `round()` function would be used.

5.  **Outputting Results:** Finally, we need to print these calculated values in a clear, descriptive way.

### What the Code Does (Summary)

This code block initializes a list of dictionaries, where each dictionary describes a specific procurement item with its name, quantity, and unit price. It then performs two main calculations:

1.  **Calculates the total quantity** of *all items combined* by summing the `quantity` field of each item in the list.
2.  **Calculates the average unit price** across the *different types of items* by summing the `price` field of each item and dividing by the total number of unique item types.

Finally, it prints both of these calculated results.

### Step-by-Step Explanation

1.  **`items = [...]`**:
    *   **Meaning:** This line creates a Python `list` named `items`. Inside this list, there are four `dictionaries`.
    *   **Reason:** This is our raw data, representing different procurement items. Each dictionary holds structured information (name, quantity, price) for one type of item. Using a list allows us to easily store and iterate over multiple items.

2.  **`total_quantity = 0`**:
    *   **Meaning:** A variable named `total_quantity` is created and initialized to `0`.
    *   **Reason:** This variable will act as an accumulator. We start it at zero because we haven't added any quantities yet, and we'll add the quantity of each item to it in the upcoming loop.

3.  **`for item in items:`**:
    *   **Meaning:** This starts a `for` loop. In each iteration of the loop, the variable `item` will successively take on the value of each dictionary from the `items` list (e.g., first it's `{"name": "Laptop", ...}`, then `{"name": "Monitor", ...}`, and so on).
    *   **Reason:** We need to process each item individually to sum their quantities. A `for` loop is the standard way to iterate over elements in a list.

4.  **`total_quantity += item["quantity"]`**:
    *   **Meaning:** Inside the loop, for the current `item` dictionary, it retrieves the value associated with the key `"quantity"` (e.g., `8` for the Laptop, `15` for the Monitor) and adds it to the `total_quantity` variable. The `+=` is shorthand for `total_quantity = total_quantity + item["quantity"]`.
    *   **Reason:** This is how we accumulate the sum of all quantities from each item into a single `total_quantity`.

5.  **`print("Total Quantity:", total_quantity)`**:
    *   **Meaning:** After the loop finishes executing for all items, this line prints the final calculated `total_quantity` to the console, preceded by the text "Total Quantity:".
    *   **Reason:** To display the result of our first calculation to the user.

6.  **`total_price = 0`**:
    *   **Meaning:** A new variable `total_price` is created and initialized to `0`.
    *   **Reason:** Similar to `total_quantity`, this variable will accumulate the sum of the prices of each *type* of item, starting from zero.

7.  **`for item in items:`**:
    *   **Meaning:** Another `for` loop that again iterates through each dictionary in the `items` list.
    *   **Reason:** We need to process each item again, this time to sum their individual prices to calculate the average price.

8.  **`total_price += item["price"]`**:
    *   **Meaning:** Inside this second loop, for the current `item` dictionary, it retrieves the value associated with the key `"price"` (e.g., `1250` for the Laptop, `320` for the Monitor) and adds it to the `total_price` variable.
    *   **Reason:** This accumulates the sum of the unit prices of all item types.

9.  **`average_price = total_price / len(items)`**:
    *   **Meaning:** This line calculates the average price. `len(items)` returns the number of items (dictionaries) in the `items` list (which is `4` in this case). The `total_price` is then divided by `4`.
    *   **Reason:** To find the average, we divide the sum of prices by the count of items that contributed to that sum.

10. **`print("Average Price per Item:", round(average_price, 2))`**:
    *   **Meaning:** This line prints the final calculated `average_price` to the console, rounded to two decimal places, preceded by the text "Average Price per Item:".
    *   **Reason:** To display the result of our second calculation, formatted for readability as prices are typically shown with two decimal places.
